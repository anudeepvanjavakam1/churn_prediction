```python
import pandas as pd
import requests
import json
```

```python
test = pd.read_csv("data/test.csv")

test.drop(columns=['CustomerID'], inplace=True)
test.head()
```

```python
test['json'] = test.to_json(orient='records', lines=True).splitlines()
test['json']
```

```python
customer = json.loads(test['json'][0])
```

```python
url = "http://127.0.0.1:9696/predict"
requests.post(url, json=customer).json()
```
